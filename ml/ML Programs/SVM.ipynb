```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.svm import SVC
```

```python
df=pd.read_csv("winequality-white.csv",sep=';')
df
```

```python
X=df.drop('quality',axis=1)
Y=df['quality']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
linear_svm=SVC(kernel='linear')
```

```python
linear_svm.fit(x_train,y_train)
```

```python
y_pred=linear_svm.predict(x_test)
```

```python
print('Linear SVM:')
print('Clssification Report:\n',classification_report(y_test,y_pred))
```

```python
con=confusion_matrix(y_test,y_pred)
sns.heatmap(con,annot=True,fmt='d',cbar=False,cmap=plt.cm.Blues)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Linear SVM')
plt.show()
```

```python
proba=linear_svm.decision_function(x_test)
plt.hist(proba,bins=50)
plt.xlabel('Predicted Quality')
plt.ylabel('Count')
plt.title('Histogram For Predicted Probability - Linear SVM')
plt.show()
```

```python
nonlinear_svm=SVC(kernel='rbf',C=10,gamma=0.1)
```

```python
nonlinear_svm.fit(x_train,y_train)
```

```python
y_pred=nonlinear_svm.predict(x_test)
```

```python
print('Non Linear SVM:')
print('Clssification Report:\n',classification_report(y_test,y_pred))
```

```python
con=confusion_matrix(y_test,y_pred)
sns.heatmap(con,annot=True,fmt='d',cbar=False,cmap=plt.cm.Blues)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Non Linear SVM')
plt.show()
```

```python
proba=linear_svm.decision_function(x_test)
plt.hist(proba,bins=50)
plt.xlabel('Predicted Quality')
plt.ylabel('Count')
plt.title('Histogram For Predicted Probability - Linear SVM')
plt.show()
```
