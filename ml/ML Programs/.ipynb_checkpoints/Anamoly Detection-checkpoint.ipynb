```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.ensemble import IsolationForest
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
X=iris.data
```

```python
clf_iso=IsolationForest(n_estimators=100,contamination=0.1,random_state=42)
```

```python
y_iso=clf_iso.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_iso,cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Outlier Detection Using Isolation Forest')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=np.where(y_iso==-1,"red","blue"),label='Isolation Forest')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Outlier Detection Using Isolation Forest')
plt.legend()
plt.show()
```

```python
clf_lof=LocalOutlierFactor(n_neighbors=20,contamination=0.1)
y_lof=clf_lof.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_lof,cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Outlier Detection Using Local Outlier Factor')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=np.where(y_lof==-1,"red","blue"),label='Local Outlier Factor')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Outlier Detection Using Local Outlier Factor')
plt.legend()
plt.show()
```
