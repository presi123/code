```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python
iris=load_iris()
X=iris.data
```

```python
pca=PCA(n_components=2)
Xpca=pca.fit_transform(X)
```

```python
dbscan_hard=DBSCAN(eps=0.5,min_samples=5)
y_hard=dbscan_hard.fit_predict(X)
```

```python
dbscan_hard=DBSCAN(eps=0.5,min_samples=5)
y_hardpca=dbscan_hard.fit_predict(Xpca)
```

```python
dbscan_soft=DBSCAN(eps=0.5,min_samples=5)
dbscan_soft.fit(X)
soft_labels=dbscan_soft.labels_
```

```python
dbscan_soft=DBSCAN(eps=0.5,min_samples=5)
dbscan_soft.fit(Xpca)
soft_labelspca=dbscan_soft.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=y_hard)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Hard DBSCAN On IRIS Dataset')
plt.show()
```

```python
plt.scatter(Xpca[:,0],Xpca[:,1],c=y_hardpca)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Hard DBSCAN And PCA On IRIS Dataset')
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=soft_labels,cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Soft DBSCAN On IRIS Dataset')
plt.show()
```

```python
plt.scatter(Xpca[:,0],Xpca[:,1],c=soft_labelspca,cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Soft DBSCAN And PCA On IRIS Dataset')
plt.show()
```
